# Challenge 5 ðŸ‘¾

```python
# Crear la clase mascota 
class Mascota:
    especie = 'gato'

# Atributo de instancia 
    def __init__(self, nombre, color):
        self.nombre = nombre 
        self.color = color 

    def hablar(self, palabra):
        print(f'cuando hablo digo {palabra}')

    def al_moverse(self, movimiento):
        print(f'Aprendi la habilidad de {movimiento}.')


gato1 = Mascota('kira', 'gris')
gato1.hablar('hola')
gato1.al_moverse('girar')
```

# Challenge 6 ðŸ‘¾

```python
# Definir la clase madre 
class AnimalesDelZoo:
    tipo = 'animal'

    # Definir los atributos de instancia de la clase madre 
    def __init__(self, nombre, especie, peso):
        self.nombre = nombre
        self.especie = especie
        self.peso = peso
#Crear la primera clase hija sin init pero con acciones diferenciadas 
class Mono(AnimalesDelZoo):

    def especialidad(self, especialidad):
        print(f'Mi especialidad es: {especialidad}')
    
    def comida_favorita(self, comida):
        print(f'Mi comida favorita es: {comida}')

# Crear la segunda clase hija pero tambien hereda a la primera clase hija (al heredar a la primera 
# clase hija tambien hereda a la clase madre)
class Leon(Mono):

    def titulo(self):
        print('Tengo el titulo de: El rey de la selva')

    
mono1 = Mono('Lucas', 'Mono', '15kg')
print(f'Hola! soy {mono1.nombre} el {mono1.especie}')
mono1.comida_favorita('Bananas')
mono1.especialidad('Trepar arboles')

print('\n')

leon1 = Leon('Zimba', 'Leon', '80kg')
print(f'Hola! soy {leon1.nombre} el {leon1.especie}')
leon1.titulo()
leon1.comida_favorita('Carne')
leon1.especialidad('cazar en campo abierto')

```

```python
# Crear una clase principal llamada personale donde podremos asignar un nombre y un saludo
class Personajes:
    tipo_de_elemento = 'Personajes'

    def __init__(self, nombre):
        self.nombre = nombre 

class Heroe(Personajes):
    rol = 'heroe'

    def presentacion(self, poder):
        print(f'Hola! Soy el heroe {self.nombre} y tengo {poder}')

class Villano(Personajes):
    rol = 'villano'

    def presentacion(self, poder):
        print(f'Hola! Soy el villano {self.nombre} y tengo {poder}')



heroe1 = Heroe('Superman')
heroe1.presentacion('super-fuerza')

print('\n')

villano1 = Villano('MisterMind')
villano1.presentacion('telekinesis')
```
