Suma: asignar dos n칰meros en dos variables diferentes cantidad_de_manzanas y cantidad_de_naranjas. Luego sumar ambas variables en cantidad_de_frutas. Imprimir la 칰ltima variable.

Resta: asignar un valor num칠rico a la variable frutas_vendidas. Luego, asignar la resta entre cantidad_de_frutas y frutas_vendidas a la variable frutas_restantes. Imprimir la 칰ltima variable.


```python
# datos de la suma
cantidad_de_manzanas = 20
cantidad_de_naranjas = 50 
cantidad_de_frutas = cantidad_de_manzanas + cantidad_de_naranjas

# resultado de la suma 
print(cantidad_de_frutas, 'frutas en total')

# datos de la resta
frutas_vendidas = 15 
frutas_restantes = cantidad_de_frutas - frutas_vendidas

# frutas vendidas
print(frutas_vendidas, 'frutas vendidas')

# resultado de la resta 
print(frutas_restantes, 'frutas restantes luego de la venta')


```


# Septimo challenge 游

```python
# Definir variables globales 
global1 = 'iPhone'
global2 = '17 Pro Max'

def mensaje_final_de_compra(global1, global2, local1, local2):
    print(f'Todo listo! La compra de su {global1} {global2} de color {local1} de {local2} de almacenamiento queda confirmada.')

# Definir la funcion con variables locales adentro 
def proceso_de_la_compra(local1, local2):
    local1 = input('Seleccione el color de su iPhone...')
    local2 = input('Seleccione el almacenamiento de su iPhone...')

    mensaje_final_de_compra(global1, global2, local1, local2)

# Por ultimo llamamos a la funcion 
proceso_de_la_compra(global1, global2)
```
